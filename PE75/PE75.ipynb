```python
def odd_1(n):
    if 1 == (n&1):
        return(True)
    else:
        return(False)
    
def odd_2(n):
    if n&1:
        return(True)
    else:
        return(False)

def odd_3(n):
    if n%2:
        return(True)
    else:
        return(False)
    
```

```python


%timeit [odd_1(x) for x in range(1000)]
```

```python
%timeit [odd_2(x) for x in range(1000)]
```

```python
%timeit [odd_3(x) for x in range(1000)]
```
