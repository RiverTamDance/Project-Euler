```python
import itertools as it
from collections import defaultdict, deque

s1 = "free"
s2 = "reef"

r = zip(s1, it.count())

D = defaultdict(deque)

for k,v in r:
    D[k].append(v)

seq = []
for l in s2:
    seq.append(D[l].popleft())


print(seq)
```
