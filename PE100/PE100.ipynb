```python
from math import isqrt


def f(x):
    d = sqrt(1+ 8*(x**2+x)) # if this is a whole number we're golden.
    #so if 1+ 8*(x**2+x) is a square we're golden.
    result = -1/2 + d/2
    return(result)

print(f(14))
print(f(84))
```
