```python
import itertools
import operator
import re

testlist = [11,22,33,44,55,66]

test1 = itertools.cycle(itertools.chain(range(1,3),range(3,4)))

[next(test1) for i in range(6)]

list(map(operator.xor,testlist,test1))

unencrypted_string = ''.join(map(chr,map(operator.xor,testlist,test1)))

result = re.search("5", unencrypted_string)

if not result ==  None:
    print(result)
```
